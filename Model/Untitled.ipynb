```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import pickle
```

```python
with open("Feature_Importance_RF_R121S_Y126F.pkl", "rb") as fp:
    d = pickle.load(fp)
```

```python
for i in range(0,5):
    print('%.4f' %d[i][1])
```

```python
plt.rcParams['font.size'] = 15
plt.rcParams['axes.linewidth'] = 1.5
```

```python
xtick_labels = []
n_feat = 10

fig, ax = plt.subplots(figsize=(12, 6))

for i in range(0,n_feat):
    ax.bar(i, d[i][1]) #, label=f'{d[i][1]:.4f}')
    #ax.bar_label(d[i][1])
    
    xtick_labels.append(d[i][0])

ax.set_xticks(np.arange(0,n_feat,1))
ax.set_xticklabels(xtick_labels, rotation='vertical', ha='right', rotation_mode='anchor')
ax.set_title("Top 10 feature importance")
ax.set_ylabel("Scores")

#ax.legend()

#plt.show()

plt.savefig('Feat_Importance_1.png', dpi=450, pad_inches=0.03, bbox_inches='tight')
```
